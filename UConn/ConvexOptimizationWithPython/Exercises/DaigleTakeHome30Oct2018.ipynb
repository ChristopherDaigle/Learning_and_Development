### Question 1: Regress lg(SGAS) on lg(PGAS).

```python
y = np.log(df['SGAS'])
x = np.log(df[['PGAS']])
```

```python
clf = LinearRegression()
clf.fit(x.values,y.values)
print('Logarithm of Gasoline Retail Price')
print(clf.coef_)
print()
print('Intercept:')
print(clf.intercept_)
```

### Is that what we are intersted in?
|**Variable**| **Defintion**                             | **Units**    |
|------------|:--------------------------------------------:|--------------:|
| SGAS       | Retail Sales Gasoline                      | 10^6 dolars  |
| PGAS       | Gasoline Retail Price                      | Cents/Gallon |

$ \widehat{log \left (SGAS\right )} = \hat\beta_{0} + \hat\beta_{1} log\left(PGAS\right)$

As we are interested in the price elasticity of gasoline consumption, this tells us the percent change in retail sales of gasoline (in 10^6 dollars) given a percent change in gasoline retail price (in cents per gallon). The interpretation of a log-log model is a percent change in y given a percent change in x (times the coefficient).

Price elasticity of demand involves the change in quantity demanded and the price of the good demanded.

**No**, there is no aspect of quantity in this model

### Question 2: Regress y on lg(PGAS).

```python
y = np.log(df['SGAS'] / df['PGAS'])
x = np.log(df[['PGAS']])
```

```python
clf = LinearRegression()
clf.fit(x.values,y.values)
print('Logarithm of Gasoline Retail Price')
print(clf.coef_)
print()
print('Intercept:')
print(clf.intercept_)
```

### Why is this misleading?
|**Variable**| **Defintion**                             | **Units**    |
|------------|:--------------------------------------------:|--------------:|
| SGAS       | Retail Sales Gasoline                      | 10^6 dolars  |
| PGAS       | Gasoline Retail Price                      | Cents/Gallon |

$ \widehat{log \left (\dfrac{SGAS}{PGAS} \right )} = \hat\beta_{0} + \hat\beta_{1} log\left(PGAS\right) = 6.0468 + 0.1951*log\left(PGAS\right) $

This is a log-log interpretation again, with Y being defined as the log of the ratio of retail sales of gasoline in 10^6 dollars to the gasoline retail price.

This is misleading as the units are not standard and the terms are in nominal and real

Statistically, this is misleading as the Y variable is composed of some of the X variable, and thus there is certainly an endogeneity issue. Mathematically, this is odd as we aren't really "solving" for anything, and if we are, we're solving it in terms of itself.

**More particularly, there is an issue of relating nominal and real values.**

### Question 3: Regress y on x2 and x3.

```python
y = np.log(df['SGAS'] / df['PGAS'])
x = df[['X2', 'X3']]
```

```python
clf = LinearRegression()
clf.fit(x.values,y.values)
print('log(INC/PALL), log(PGAS/PALL)')
print(clf.coef_)
print()
print('Intercept:')
print(clf.intercept_)
```

### Provide some motivation for this choice.
|**Variable**| **Defintion**                             | **Units**    |
|------------|:--------------------------------------------:|--------------:|
| SGAS       | Retail Sales Gasoline                      | 10^6 dolars  |
| PGAS       | Gasoline Retail Price                      | Cents/Gallon |
| INC        | Nominal Disposable Income                  | 10^9 dollars |
| PALL       | $\dfrac{1982 + 1983 +1984}{3} = 100$       |              |

$ \widehat{log \left (\dfrac{SGAS}{PGAS} \right )} = \hat\beta_{0} + \hat\beta_{1} log \left (\dfrac{INC}{PALL} \right ) + \hat\beta_{2} log \left (\dfrac{PGAS}{PALL} \right ) = 4.9860 + 0.5732 * log \left (\dfrac{INC}{PALL} \right ) + -0.5276 * log \left (\dfrac{PGAS}{PALL} \right )$

The choices of real-income and real-gas price may be a good choice for estimating the elasticity of quantity demanded as economic theory tells us that only real opportunities matter (i.e. relative prices to relative goods, homogeneity of degree zero and all that) and that demand is a function of prices and wealth.

The results seem reasonable: as wealth increases, demand increases (shifting quantity demanded, which is what's being observed), as price increases, quantity demanded decreases.
